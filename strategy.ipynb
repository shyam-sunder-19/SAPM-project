```python
import pandas as pd
import math
import numpy as np
```

```python
coforge = pd.read_csv('coforge.csv')
hcl = pd.read_csv('hcl.csv')
infosys = pd.read_csv('infosys.csv')
mindtree = pd.read_csv('mindtree.csv')
tcs= pd.read_csv('tcs.csv')
oracle = pd.read_csv('oracle.csv')
techm = pd.read_csv('techm.csv')
mphasis = pd.read_csv('mphasis.csv')
wipro = pd.read_csv('wipro.csv')

stocks = [coforge, hcl, infosys, tcs, oracle, techm, mindtree, mphasis, wipro]
```

```python
def calculate_sharpe(returns, rf):
    
    mean = returns.mean(axis=1)
    std = returns.std(axis=1)
    
    return (mean - rf)/std
```

```python
def rebalance_vol(stk, rebal_period):
    
    vol_helper = pd.DataFrame()
    weights = pd.DataFrame()
    
    for j in range(0, 1782, rebal_period):
        
        arr = []
        for i in range(len(stk)):
            arr.append(stk[i]['volatility '+str(math.floor(rebal_period/30))][j])
            
        vol_helper['vol'] = arr

        vol_helper['inv_vol'] = 1/vol_helper['vol']

        sum_inv_vol = vol_helper['inv_vol'].sum()

        vol_helper['weights'] = vol_helper['inv_vol']/sum_inv_vol
        
        weights[str(j)] = vol_helper['weights'] 
    
    return weights

weights = rebalance_vol(stocks, 90)
weights
```

```python
def rebalance_mom(stk, rebal_period):
    mom_helper = pd.DataFrame()
    weights = pd.DataFrame()
    
    for j in range(0, 1782, rebal_period):
        
        arr = []
        mom1 = []
        for i in range(len(stk)):
            arr.append(stk[i]['momentum '+str(math.floor(rebal_period/30))][j])
            
        mom_helper['mom'] = arr
        for mom in arr:
            mom1.append(max(0, mom))
        
        mom_helper['mom_1'] = mom1

        sum_mom_1 = mom_helper['mom_1'].sum()

        mom_helper['weights'] = mom_helper['mom_1']/sum_mom_1
        
        weights[str(j)] = mom_helper['weights'] 
    
    return weights

weights = rebalance_mom(stocks, 90)
weights
```

```python
def rebalance_momvol(stk, rebal_period):
    helperdf = pd.DataFrame()
    weights = pd.DataFrame()
    
    for j in range(0, 1782, rebal_period):
        mom = []
        vol = []
        for i in range(len(stk)):
            mom.append(stk[i]['momentum '+str(math.floor(rebal_period/30))][j])
            vol.append(stk[i]['momentum '+str(math.floor(rebal_period/30))][j])
        
            
        helperdf['mom'] = mom
        helperdf['vol'] = vol
        
        vol1 = helperdf.iloc[:][helperdf['mom'] > 0]['vol']
        vol1_sum = vol1.sum()
        vol_weights = vol1/vol1_sum
        
        helperdf['weights'] = 0

        indices = helperdf.iloc[:][helperdf['mom'] > 0].index

        for ind, wht in zip(indices, vol_weights):
            helperdf.loc[ind, 'weights'] = wht
            
        weights[str(j)] = helperdf['weights']
            
    return weights
        
rebalance_momvol(stocks, 90)     
```

```python
def calc_returns(weights, stk, rebal_period):
    rets = pd.DataFrame()
    gains = pd.DataFrame()
    returns = pd.DataFrame()
    
    for j in range(rebal_period, 1711, rebal_period):
        ret = []
        for i in range(len(stk)):
            ret.append((stk[i]['Close'][j] - stk[i]['Close'][j-rebal_period])/stk[i]['Close'][j-rebal_period])
            
        rets[str(j)] = ret
        
        gains = rets[str(j)] * weights[str(j-rebal_period)]
    
        returns.loc[0,str(j-rebal_period)] = gains.sum()
    
    product = 1
    a = returns + 1
    for b in a.columns:
        product *= a[b]
    
    sharpe = calculate_sharpe(returns, 0.070783199)
    
    return returns, product, sharpe

calc_returns(rebalance_mom(stocks, 90), stocks, 90)
```

```python
calc_returns(rebalance_momvol(stocks, 90), stocks, 90)
```
